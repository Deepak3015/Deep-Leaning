```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# list of height
height = [150,160,165,185]
# list of weight
weight = [70, 80, 90, 100]
```

```python
plt.plot(height,weight)
plt.xlabel("height")
plt.ylabel("weight")
```

```python
calories_burnt = [65, 75, 95, 99]
```

```python
# draw the plot for calories burnt
plt.plot(calories_burnt)
# draw the plot for weight
plt.plot(weight)
plt.xlabel("Caliries")
plt.ylabel("Weight")
```

```python
plt.plot(calories_burnt)
plt.plot(weight)
plt.xlabel('Caliers')
plt.ylabel('Person')
plt.xticks(ticks = [0,1,2,3,],labels = ['P1',"P2",'P3',"P4"]);
```

```python
plt.figure(figsize=(15,7))
plt.plot(calories_burnt,marker = 'x')
plt.plot(weight,"y--",marker = 'o')
plt.xlabel('Caliers')
plt.ylabel('Person')
plt.legend(labels =["Cailiers","Weight"],loc ='lower right' )
plt.xticks(ticks = [0,1,2,3],labels = ["P1","P2","P3","P4"])
```
